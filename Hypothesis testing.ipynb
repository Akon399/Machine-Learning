```python
import numpy as np 
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
mu = 0 # mean.
sigma = 1 # std deviation.
alpha = 0.05 # significance value.
```

```python
# Generate the x values & lets plot the normal curve.

x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
plt.plot(x, norm.pdf(x, mu, sigma), label = 'Normal Distribution')
```

```python
# Calculate the critical value(one-tailed test).

critical_value = norm.pdf(1 - alpha)
```

```python
# Fill the critical region.

plt.fill_between(x, 0, norm.pdf(x, mu, sigma), where=(x>=critical_value), color='red')
```

```python
# Test statistic.

test_statistic = 1.8
```

```python
# Plot the test stats.

plt.axvline(test_statistic, color='green', linestyle = 'dashed', linewidth=2, label='Test Stats')
```

```python
# Annotations 

plt.annotate('Critical value', xy=(critical_value, 0.1), xytext=(critical_value+1,0.15),
             arrowprops=dict(facecolor='black', shrink=0.05))
```

```python
# Titles & labels.

plt.title('Positive hypothesis test result')
plt.xlabel('Value')
plt.ylabel('Probability density')
plt.legend()
```

```python
plt.show()
```
