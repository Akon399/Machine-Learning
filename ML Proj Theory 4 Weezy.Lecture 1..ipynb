```python
# Date 16 Jan 2023.
# Quick Theory & disscusion.



# Q1 What is Machine Learning?

# ---> Subset of AI that refers to a set of Mathematical/Statistcal Algorithms & models to learn from data 
#      without being programmed.



# Q2 Types of Machine Learning & there algorithms.

# ---> 2.1.Supervised Learning.
#          SL Algorithm here gives a robot an ability to Predict anything with high intelligence.
#          SL algorithm is task driven, meaning it can Classify(identify fraud)/Predict(tell us weather tommorrow).
#          SL = Regression(R) & Classification(C).
#          SL Algorithms = KNN(C:Approximation), Linear Regression(R:Predicts desired output), Naive Bayes(C:Probabilisitc model).

# ---> 2.2.Unsupervised Learning.
#          UL algorithms are your Recommender Systems used in Youtube, Spotify, Netflix etc.
#          Target Marketing uses UL Algorithms.
#          UL learns with unlabled data, meaning you just give it data & it figures out the rest.
#          UL Algorithms = Gaussian Mixture, Heirarchical Clustering, K-Means Clustering.

# ---> 2.3.Reinforcement Learning.
#          Real time decision making.
#          Game AI like AlphaGo,a compter programme built by google & it plays games(Chess,GO,Cards etc) at a super human level.
#          Robot Navigation like Self Drving Cars they use RL Alg.


# Q3 How does ML work?

# ---> We must build a engine/ model that we must give lots & lots of data(pics of balls), so that it can learn from & create patterns.
#      The patterns will help the machine tell us if this is a soccer ball/golf ball, depending on problem where solving.
#      If we can give our machine a basket ball, the machine will tell us that its neither, since we have never trained it 
#      on basket balls.




# Q4 What steps do you we take in a ML project?

# ---> 1.Gather & Import data.

# ---> 2.Data Cleaning.
#        Remove duplicates, outliers & deal with Null spaces.
#        Normalisation.
#        Unclean data will result in poor perf & wrong outputs from our model.

# ---> 3.Split the data into Training(80%) & Testing(20%).

# ---> 4.Model creation & selection.
#        By model we refer to Algorithms, theres tons of Algorithms to use such as Decision Trees,SVM,KNN etc.
#        Each model gives a different accuracy score, it all depends on a problem.

# ---> 5.Train the model.

# ---> 6.Make Predictions.
#        Most of the time, predictions are not accurate, so we need step 7.

# ---> 7.Evaluate & Improve.
#        We do this by measuring Accuracy.
#        We might select & replace an existing Algorithm with another Algorithms, so we can improve our Acc score.

# ---> 8.Deploy & Maintain our model.
#        We will cover this in the future, as it needs Cloud Computing knowledge.But deploying is the last step.



# Where is Machine Learning useful? 

# ---> Robotics, Self driving cars, Forecasting the weather & stockmarket, Language Processing & many many more etc.
#      Machine Learning is at the fore front of of Advancing these feilds.


```
