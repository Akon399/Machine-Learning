```python
sns.heatmap(machine_scf.isnull(), cbar=False, cmap='viridis')
plt.title('SCF(Flexible scope) : SAP Database.')
plt.show()
```

```python
machine_scf = machine_scf.drop(['20.5 System_Status','20.6 User_Status'], axis=1)
numeric_cols = machine_scf.select_dtypes(include='number').columns
machine_scf[numeric_cols] = machine_scf[numeric_cols].fillna(machine_scf[numeric_cols].mean())

# Fill missing values for categorical columns with the mode
categorical_cols = machine_scf.select_dtypes(include='object').columns
machine_scf[categorical_cols] = machine_scf[categorical_cols].fillna(machine_scf[categorical_cols].mode().iloc[0])

# machine_scf.isnull().sum()
```

## How to choose an algorithm for machine failure?
### Is it A or B? ---> Classification algorithm. Think Categorical data.
### Is it weird? ---> Detection algorithm. Think a model that detects scams in credit card, algorithm must check pattern.
### How much/ How many? ---> Regression algorithm. What will be the temparature tomorrow?
### How is this organised? ---> Clustering algorithm. Think grouping the same points in your data togehter.
### What should i do next? ---> Reinforcement algorithm. Let a computer decide, & then you correct the computer afterwards.
